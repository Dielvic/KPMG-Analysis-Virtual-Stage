```python
#Remplissage de la colonne catégorie_
df["Catégorie_"]= "[NL] - AUTRES"
df.loc[df["Nom_e-mail"].str.contains("NL-SERVICES"), 'Catégorie_'] = '[NL] - SERVICE'
df.loc[(df["Nom_e-mail"].str.contains("OFFRE")|df["Nom_e-mail"].str.contains("OAAE") |df["Nom_e-mail"].str.contains("ETRADI")) , 'Catégorie_'] = "[NL] - OFFRE"
df.loc[df["Nom_e-mail"].str.contains("LABO"), 'Catégorie_'] = "[NL] - LABO"
df.loc[(df["Nom_e-mail"].str.contains("VERZENIOS")& df["Envois_"]>1), 'Catégorie_'] = "[NL] - LABO"
df.loc[df["Nom_e-mail"].str.contains("NL-ETRADI"), 'Catégorie_'] = "[NL] - OFFRE"
df.loc[df["Nom_e-mail"].str.contains("INFOFLASH") , 'Catégorie_'] = "[NL] - INFO"
df.loc[df["Nom_e-mail"].str.contains("IF"), 'Catégorie_'] = "[NL] - INFO"
df.loc[df["Nom_e-mail"].str.contains("EVENT"), 'Catégorie_'] = "[NL] - EVENT"
```

```python
Type = [["[AUTRES] - AUTRES", "[AUTRES] - HEBDOCP"],["[EVENT] - AUTRES", "[EVENT] - CARREFOURS DE L'INSTALLATION","[EVENT] - CONF BUXERAUD",
        "[EVENT] - INTERFIMO", "[EVENT] - PHARMAGORA", "[EVENT] - RENCONTRES DE L'OFFICINE"] , ["[INFO] - INFO LABO", "[INFO] - INFOFLASH", 
        "[INFO] - INFOPRO", "[INFO] - INFOREGLEMENTAIRE", "[INFO] - COVID"], ["[LABO] - CONTENU", "[LABO] - INFO INSTIT", "[LABO] - INFO PRODUIT",
        "[LABO] - OFFRE"], ["[LINK] - MON COMPTE OCP", "[OCPF] - VISIO FORMATION"], ["[OFFRES] - AUTRES OFFRES", "[OFFRES] - COVID", "[OFFRES] - EQUIPEMENT",
        "[OFFRES] - ETRADI", "[OFFRES] - GENERIQUES", "[OFFRES] - GRIPPE", "[OFFRES] - INCONTOURNABLES", "[OFFRES] - MARQUE CONSEIL",
        "[OFFRES] - OAAE", "[OFFRES] - OCPH", "[OFFRES] - OFFRE LINK", "[OFFRES] - OFFRES DU MOIS", "[OFFRES] - SOLUTIONS ACHATS",
        "[OFFRES] - TEMPS FORT PROMOTIONNEL"], ["[SERVICES] - AUTRES SERVICES", "[SERVICES] - DATAE", "[SERVICES] - FINANCEMENT", "[SERVICES] - FINANCEMENT EVENT",
        "[SERVICES] - FINANCEMENT EXTENSIO", "[SERVICES] - FINANCEMENT INITIO", "[SERVICES] - LINK", "[SERVICES] - MON OCPH", "[SERVICES] - OCPF",
        "[SERVICES] - OCPMFO", "[SERVICES] - PACK SERVICE", "[SERVICES] - PETITES ANNONCES", "[SERVICES] - SERVICE RELATION CLIENT",
        "[SERVICES] - SERVICES DE GESTION"]]
```

```python
df.loc[df["Catégorie_"].str.endswith("AUTRES") & df["Nom_e-mail"].str.contains("HEBDOCP") , 'type_'] = Type[0][1]
df.loc[df["Catégorie_"].str.endswith("AUTRES") & df["Nom_e-mail"].str.contains("AUTRES") , 'type_'] = Type[0][0]
df.loc[df["Catégorie_"].str.endswith("EVENT") & df["Nom_e-mail"].str.contains("AUTRES") , 'type_'] = Type[1][0]
df.loc[df["Catégorie_"].str.endswith("EVENT") & df["Nom_e-mail"].str.contains("CARREFOURS DE L'INSTALLATION") , 'type_'] = Type[1][1]
df.loc[df["Catégorie_"].str.endswith("EVENT") & df["Nom_e-mail"].str.contains("CONF BUXERAUD") , 'type_'] = Type[1][2]
df.loc[df["Catégorie_"].str.endswith("EVENT") & df["Nom_e-mail"].str.contains("INTERFIMO") , 'type_'] = Type[1][3]
df.loc[df["Catégorie_"].str.endswith("EVENT") & df["Nom_e-mail"].str.contains("PHARMAGORA") , 'type_'] = Type[1][4]
df.loc[df["Catégorie_"].str.endswith("EVENT") & df["Nom_e-mail"].str.contains("RENCONTRES DE L'OFFICINE") , 'type_'] = Type[1][5]
df.loc[df["Catégorie_"].str.endswith("INFO") & df["Nom_e-mail"].str.contains("INFO LABO") , 'type_'] = Type[2][0]
df.loc[df["Catégorie_"].str.endswith("INFO") & df["Nom_e-mail"].str.contains("INFOFLASH") , 'type_'] = Type[2][1]
df.loc[df["Catégorie_"].str.endswith("INFO") & df["Nom_e-mail"].str.contains("INFOPRO") , 'type_'] = Type[2][2]
df.loc[df["Catégorie_"].str.endswith("INFO") & df["Nom_e-mail"].str.contains("INFOREGLEMENTAIRE") , 'type_'] = Type[2][3]
df.loc[df["Catégorie_"].str.endswith("INFO") & df["Nom_e-mail"].str.contains("COVID") , 'type_'] = Type[2][4]
df.loc[df["Catégorie_"].str.endswith("LABO") & df["Nom_e-mail"].str.contains("CONTENU") , 'type_'] = Type[3][0]
df.loc[df["Catégorie_"].str.endswith("LABO") & df["Nom_e-mail"].str.contains("INFO INSTIT") , 'type_'] = Type[3][1]
df.loc[df["Catégorie_"].str.endswith("LABO") & (df["Nom_e-mail"].str.contains("INFO PRODUIT") | df["Nom_e-mail"].str.contains("LILLY") | df["Nom_e-mail"].str.contains("VERZENIOS")), 'type_'] = Type[3][2]
df.loc[df["Catégorie_"].str.endswith("LABO") & df["Nom_e-mail"].str.contains("OFFRE") , 'type_'] = Type[3][3]
df.loc[df["Catégorie_"].str.endswith("LINK") & df["Nom_e-mail"].str.contains("MON COMPTE OCP") , 'type_'] = Type[4][0]
df.loc[df["Catégorie_"].str.endswith("OCPF") & df["Nom_e-mail"].str.contains("VISIO FORMATION") , 'type_'] = Type[4][1]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & df["Nom_e-mail"].str.contains("AUTRES OFFRES") , 'type_'] = Type[5][0]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & df["Nom_e-mail"].str.contains("COVID") , 'type_'] = Type[5][1]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & (df["Nom_e-mail"].str.contains("EQUIPEMENT")| df["Nom_e-mail"].str.contains("EQT")) , 'type_'] = Type[5][2]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & df["Nom_e-mail"].str.contains("ETRADI") , 'type_'] = Type[5][3]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & df["Nom_e-mail"].str.contains("GENERIQUES") , 'type_'] = Type[5][4]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & (df["Nom_e-mail"].str.contains("GRIPPE") | df["Nom_e-mail"].str.contains("GEN")) , 'type_'] = Type[5][5]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & df["Nom_e-mail"].str.contains("INCONTOURNABLES") , 'type_'] = Type[5][6]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & (df["Nom_e-mail"].str.contains("MARQUE CONSEIL") | df["Nom_e-mail"].str.contains("MC")) , 'type_'] = Type[5][7]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & df["Nom_e-mail"].str.contains("OAAE") , 'type_'] = Type[5][8]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & df["Nom_e-mail"].str.contains("OCPH") , 'type_'] = Type[5][9]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & df["Nom_e-mail"].str.contains("OFFRE LINK") , 'type_'] = Type[5][10]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & (df["Nom_e-mail"].str.contains("ODM") | df["Nom_e-mail"].str.contains("LILLY")), 'type_'] = Type[5][11]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & df["Nom_e-mail"].str.contains("SOLUTIONS ACHATS") , 'type_'] = Type[5][12]
df.loc[df["Catégorie_"].str.endswith("OFFRE") & df["Nom_e-mail"].str.contains("TEMPS FORT PROMOTIONNEL") , 'type_'] = Type[5][13]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & df["Nom_e-mail"].str.contains("AUTRES SERVICES") , 'type_'] = Type[6][0]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & df["Nom_e-mail"].str.contains("DATAE") , 'type_'] = Type[6][1]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & df["Nom_e-mail"].str.contains("FINANCEMENT") , 'type_'] = Type[6][2]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & df["Nom_e-mail"].str.contains("FINANCEMENT EVENT") , 'type_'] = Type[6][3]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & df["Nom_e-mail"].str.contains("FINANCEMENT EXTENSIO") , 'type_'] = Type[6][4]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & df["Nom_e-mail"].str.contains("INITIO") , 'type_'] = Type[6][5]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & (df["Nom_e-mail"].str.contains("LINK") | df["Nom_e-mail"].str.contains("BONSUPGRADE")), 'type_'] = Type[6][6]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & df["Nom_e-mail"].str.contains("OCPH") , 'type_'] = Type[6][3]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & df["Nom_e-mail"].str.contains("OCPF") , 'type_'] = Type[6][2]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & (df["Nom_e-mail"].str.contains("OCPMFO") | df["Nom_e-mail"].str.contains("MFO")) , 'type_'] = Type[6][9]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & df["Nom_e-mail"].str.contains("PACK SERVICE") , 'type_'] = Type[6][10]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & df["Nom_e-mail"].str.contains("PETITES ANNONCES") , 'type_'] = Type[6][11]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & df["Nom_e-mail"].str.contains("SERVICE RELATION CLIENT") , 'type_'] = Type[6][12]
df.loc[df["Catégorie_"].str.endswith("SERVICE") & (df["Nom_e-mail"].str.contains("SERVICES DE GESTION") | df["Nom_e-mail"].str.contains("SG") 
                                                   | df["Nom_e-mail"].str.contains("SERVICESDEGESTION") | df["Nom_e-mail"].str.contains("DEMATVAGUE")) , 'type_'] = Type[6][13]
```
