## II-B-1 Preprocessig

######  outliers Management

```python
#function to find outliers
def iqr_outliers(df):
    out=[]
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3-q1
    Lower_tail = q1 - 1.5 * iqr
    Upper_tail = q3 + 1.5 * iqr
    for i in df:
        if i > Upper_tail or i < Lower_tail :
            out.append(i)
    return out
```

```python
d =iqr_outliers(dataset['Age'])
d.sort()
d[0]
```
